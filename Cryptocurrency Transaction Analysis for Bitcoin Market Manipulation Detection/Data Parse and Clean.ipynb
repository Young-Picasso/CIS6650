```python
def import_csv_to_dataframe(file):
    df = pd.read_csv(file)
    return df
```

```python
def remove_duplicates(df):
    df = df.drop_duplicates()
    return df
```

```python
def drop_irrelevant_data(df):
    df = df[['Trade_Id', 'Date', 'User_Id', 'Type', 'Currency', 'Bitcoins', 'Money']]
    return df
```

```python
def get_trades_per_trade_id(df):
    return dict(df.groupby('Trade_Id').size())
```

```python
def get_single_row_transactions(trades_per_id):
    single_row_transactions = []
    
    for key, value in trades_per_id.items():
        if value == 1:
            single_row_transactions.append(key)
            
    return single_row_transactions
```

```python
def drop_single_row_transactions(df):
    df = df[df.groupby('Trade_Id').Trade_Id.transform('count') >1]
    return df
```

```python
def save_0_duplicate_rows_df_to_csv(df, file):
    df.to_csv(file)
```

```python
for _, _, files in os.walk(path):
    for file in files:
        print(file)
        df = import_csv_to_dataframe(str(file))
        df = remove_duplicates(df)
        df = drop_irrelevant_data(df)
        df = drop_single_row_transactions(df)
        save_0_duplicate_rows_df_to_csv(df, file)
```

```python
def filter_rows_by_values(df, col, values):
    return df[~df[col].isin(values)]
```
